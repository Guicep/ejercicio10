Utilizo tres bibliotecas para la resolucion de mi problema, string para la manipulacion de los datos, os y sys para la ubicacion de los archivos

```python
import os
import sys
import string
```

PARTE 1:
Utilizo .readlines() para que me devuelva una lista de todos los elementos que contienen los archivos y los guardo en una variable<br>
<br>
PARTE 2:
Creo list_total_score para guardar la suma de ambas notas y creo un diccionario como mi estructura para guardar las listas ya que puedo utilizar el indice como una categoria<br>
<br>
PARTE 3:
Leyendo los elementos de las listas utilizo (string.punctuation + " \n") para remover sus caracteres innecesarios,<br>
luego con los datos de las notas las convierto a entero para asi realizar la suma y guardarlas en list_total_score,<br>
guardo las nuevas listas en mi diccionario y lo retorno
<br>

```python
def get_student_structure(arch1, arch2, arch3):
    #PARTE 1
    with (open(arch1, 'r', encoding="UTF-8") as file_names, 
            open(arch2, 'r', encoding="UTF-8") as file_1_scores, 
            open(arch3, 'r', encoding="UTF-8") as file_2_scores): 
        names = file_names.readlines()
        first_scores = file_1_scores.readlines()
        second_scores = file_2_scores.readlines()
    #PARTE 2
    total_score = []
    my_dictionary = {"Alumnos": list, "Promedio": list}
    #PARTE 3
    for element_names, element_1_score, element_2_score, index in zip(names, first_scores, second_scores, range(len(names))):
        element_names = "".join(char for char in element_names if char not in (string.punctuation + " \n"))
        names[index] = element_names

        element_1_score = int("".join(char for char in element_1_score if char not in (string.punctuation + " \n")))
        element_2_score = int("".join(char for char in element_2_score if char not in (string.punctuation + " \n")))
        total_score.append(element_1_score + element_2_score)
            
    my_dictionary["Alumnos"] = names
    my_dictionary["Promedio"] = total_score
    return my_dictionary
```

PARTE 4:
Determino el promedio de todas las notas guardadas en el diccionario con una funcion <br>
<br>
PARTE 5:
Y en la misma retorno los alumnos que estan por debajo de ese promedio con una lista

```python
def get_average(dictionary):
    #PARTE 4
    add = 0
    for element_number in dictionary["Promedio"]:
        add+= element_number
    average = add / len(dictionary["Promedio"])
    #PARTE 5
    below_average = []
    for element_student, element_number in zip(dictionary["Alumnos"], dictionary["Promedio"]):
        if element_number < average:
            below_average.append(element_student)

    return average, below_average
```

Declaro las ubicaciones de mis archivos utilizando la libreria os + sys y utilizo las funciones declaradas anteriormente,<br>
al final imprimo todos los nombres de los alumnos por debajo del promedio.

```python

my_arch1 = os.path.join(sys.path[0] + r"\nombres_1.txt")
my_arch2 = os.path.join(sys.path[0] + r"\eval1.txt")
my_arch3 = os.path.join(sys.path[0] + r"\eval2.txt")

my_dictionary = get_student_structure(my_arch1, my_arch2, my_arch3)
aver, below = get_average(my_dictionary)
print(f"Los siguientes alumnos tienen notas por debajo del promedio {aver}: ", *below, sep="\n")
```
